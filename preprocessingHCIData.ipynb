```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("HCICollatedData.csv")
df.head(100)
```

```python
df['tab_nopic_clicks'] = df['tab-noPic-first-clicks'] + df['tab-noPic-second-clicks']
df['list_nopic_clicks'] = df['list-noPic-first-clicks'] + df['list-noPic-second-clicks']
df['paragraph_nopic_clicks'] = df['paragraph-noPic-first-clicks'] + df['paragraph-noPic-second-clicks']
df['tab_pic_clicks'] = df['tab-pic-first-clicks'] + df['tab-noPic-second-clicks']
df['list_pic_clicks'] = df['list-pic-first-clicks'] + df['list-pic-second-clicks']
df['paragraph_pic_clicks'] = df['paragraph-pic-first-clicks'] + df['paragraph-pic-second-clicks']
df
```

```python
columns_to_check = ['tab_nopic_clicks', 'list_nopic_clicks', 'paragraph_nopic_clicks', 
                    'tab_pic_clicks', 'list_pic_clicks', 'paragraph_pic_clicks']

for index, row in df.iterrows():
    participant = row['Participant']
    for column in columns_to_check:
        if row[column] != 0:
            df.at[index, column] = 1

df
```

```python
columns_to_check = ['tab_nopic_clicks', 'list_nopic_clicks', 'paragraph_nopic_clicks', 'tab_pic_clicks', 'list_pic_clicks', 'paragraph_pic_clicks']

new_rows = []
for index, row in df.iterrows():
    participant = row['Participant']
    for column in columns_to_check:
        if row[column] != 0:
            new_row = {'Participant': participant, column: 1}
            new_rows.append(new_row)

result_df = pd.DataFrame(new_rows)

result_df.fillna(0, inplace=True)
```

```python
df = pd.read_csv("HCICollatedData.csv")
merged = pd.merge(df, result_df, on='Participant', how='outer')
merged
```
