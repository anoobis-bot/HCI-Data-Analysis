```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("HCICollatedData.csv")
```

```python
num_participants = 20
num_columns = 7 

participants = ['P' + str(i) for i in range(1, 11)] + ['S' + str(i) for i in range(1, 11)]

data = []
for participant in participants:
    for i in range(num_columns):
        row = [0] * num_columns
        row[i] = 1
        row[0] = participant
        data.append(row)

column_names = ['Participant'] + ['tab_nopic_clicks', 'list_nopic_clicks',
                                  'paragraph_nopic_clicks', 'tab_pic_clicks',
                                  'list_pic_clicks', 'paragraph_pic_clicks']

result_df = pd.DataFrame(data, columns=column_names)

result_df = result_df[(result_df.iloc[:, 1:] != 0).any(axis=1)]
```

```python
result_df
```

```python
merged = pd.merge(df, result_df, on='Participant', how='outer')
merged
```

```python
columns_to_convert = merged.columns.difference(['Participant'])
merged[columns_to_convert] = merged[columns_to_convert].apply(pd.to_numeric, errors='coerce')
merged
```

```python
mask = merged['tab_nopic_clicks'] == 1
merged.loc[mask, 'sumClicks'] = merged.loc[mask, 'tab-noPic-first-clicks'] + merged.loc[mask, 'tab-noPic-second-clicks']
merged.loc[mask, 'avgTime'] = (merged.loc[mask, 'tab-noPic-first-time'] + merged.loc[mask, 'tab-noPic-second-time'])/2
merged.loc[mask, 'clicks1'] = merged.loc[mask, 'tab-noPic-first-clicks']
merged.loc[mask, 'clicks2'] = merged.loc[mask, 'tab-noPic-second-clicks']
merged.loc[mask, 'time1'] = merged.loc[mask, 'tab-noPic-first-time']
merged.loc[mask, 'time2'] = merged.loc[mask, 'tab-noPic-second-time']
```
